```python
import pandas as pd
import numpy as np
from math import nan
```

```python
week3 = pd.read_csv('nfloffenseweek3.csv')
week4 = pd.read_csv('nfloffenseweek4.csv')
week5 = pd.read_csv('nfloffenseweek5.csv')
life_expct = pd.read_csv('life_expectancy.csv')
pop_den = pd.read_csv('population_density.csv')
qulty_lf = pd.read_csv('quality_of_life.csv')
hght_wght_data = pd.read_csv('height_weight_data.csv')
iq = pd.read_csv('iq.csv')
```

3. Make up 1 interesting question about this dataset. Answer it.

```python
# What ten teams had the highest percentage of socring drives, then 1st downs, then total points scored over Week 3 - Week 5
from functools import reduce
week3data = week3.sort_values(["percentage_scoring_drives", "1st_downs","points_scored"], ascending= False)
week4data = week4.sort_values(["percentage_scoring_drives", "1st_downs","points_scored"], ascending= False)
week5data = week5.sort_values(["percentage_scoring_drives", "1st_downs","points_scored"], ascending= False)

#following code found at https://www.statology.org/pandas-merge-multiple-dataframes/
dfs = [week3data,week4data,week5data]
merged_df = reduce(lambda  left,right: pd.merge(left,right,on=['team'], how='outer'), dfs)

merged_df.head(10)


```

## World Population Data

4. Join the height and wieght data to the iq data

```python
iq.merge(hght_wght_data, how='left', on='country')
```

5. Join population density to quality of life data

```python
pop_den.merge(qulty_lf, how='left',on='country')
```

6. Join life expectancy to quality of life data

```python
life_expct.merge(qulty_lf,how='left',on='country')
```

7. Join iq, life expectancy, quality of life, and population density into 1 dataframe

```python
joinedDF1 = iq.merge(life_expct, on = 'country')
joinedDF2 = qulty_lf.merge(pop_den, on = 'country')

joinedDF1.merge(joinedDF2, how='outer',on='country')
```

9. Make up 1 interesting question about one of these datasets and answer it.

```python
#For each country, what is the average life expectancy 

final_merge = qulty_lf.merge(life_expct, on='country')

life = ['male_life_expectancy','female_life_expectancy']

final_merge['Total_Average_Life_Expectancy'] = final_merge[life].mean(axis=1)

saftestDF = final_merge.sort_values('safety', ascending=False)

print(saftestDF[['country','Total_Average_Life_Expectancy']].to_string(index=True))
```

10. Make up 1 interesting question about one of these datasets and answer it.

```python
#What is the average life expectancy of the 20 most populated countries

final_merge = pop_den.merge(life_expct, on='country')

pd.to_numeric(pop_den.population, errors='coerce')

life = ['male_life_expectancy','female_life_expectancy']

final_merge['Total_Average_Life_Expectancy'] = final_merge[life].mean(axis=1)

final_merge['population'] = final_merge['population'].str.replace("M","")
final_merge['population'] = final_merge['population'].str.replace(",","")

final_merge.population = pd.to_numeric(final_merge.population)

True_saftestDF = final_merge.sort_values('population', ascending=False).head(20)

True_saftestDF.columns = True_saftestDF.columns.str.replace('population','population_in_millions')

print(True_saftestDF[['country','Total_Average_Life_Expectancy','population_in_millions']].to_string(index=True))
```
